```
import pandas as pd
data = pd.read_csv("/content/final_data.csv")
data.head()
```

```
values = data['Annotation 1'].value_counts().keys().tolist()
counts = data['Annotation 1'].value_counts().tolist()
```

```
import matplotlib.pyplot as plt
colors=['blue', 'yellow', 'green', 'orange', 'pink', 'salmon', 'lightseagreen', 'magenta', 'aliceblue']
plt.pie(counts,labels=values, colors=colors, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
